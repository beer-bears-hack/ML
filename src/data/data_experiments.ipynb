```python
cols = ['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']  # нужные колонки

mask = cte_df[cols].notna().any(axis=1)
error_cte = cte_df[mask]
error_cte
```

```python
rows_to_drop = error_cte["Идентификатор СТЕ"].to_list()
cte_df = cte_df[~cte_df["Идентификатор СТЕ"].isin(rows_to_drop)]

for col in ['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10']:
    print(cte_df[col].unique())
```

```python
cte_df.drop(columns=['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], inplace=True)
```

```python
cte_df = cte_df.rename(columns={
    'Идентификатор СТЕ': "CTE_id", 
    'Наименование СТЕ': "CTE_name",
    'Категория': "category",
    'Производитель': "manufacturer",
    'характеристики СТЕ': "characteristics"
})
```

```python
cte_nan = cte_df[cte_df.isna().any(axis=1)]
cte_nan
```

```python
for cte_id in [39114363, 36715295]:
    mask = cte_df["CTE_id"] == cte_id
    cte_df.loc[mask, "characteristics"] = (
        cte_df.loc[mask, "manufacturer"]
        .astype(str)
        .apply(lambda x: x.split("\t")[1])
    )
    cte_df.loc[mask, "manufacturer"] = (
        cte_df.loc[mask, "manufacturer"]
        .astype(str)
        .apply(lambda x: x.split("\t")[0])
    )
```

```python
cte_df.dropna(inplace=True)
print(cte_df.shape)
```

```python
print(f"Duplicates: {cte_df.shape[0] - cte_df['CTE_id'].nunique()}")
```

```python
cte_df.head()
cte_df.to_csv("../../data/prep/CTE.csv", index=False)
```
